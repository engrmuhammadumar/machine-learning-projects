```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the sine wave
frequency = 3       # Frequency in Hz
amplitude = 1       # Amplitude
time_end = 1        # End time in seconds
sampling_rate = 1000  # Sampling rate in Hz

# Generate the time array
t = np.linspace(0, time_end, int(time_end * sampling_rate), endpoint=False)

# Generate the sine wave
y = amplitude * np.sin(2 * np.pi * frequency * t)

# Number of Fourier series terms (harmonics)
n_terms = 10

# Initialize arrays for Fourier coefficients
a0 = np.mean(y)
an = np.zeros(n_terms)
bn = np.zeros(n_terms)

# Calculate the Fourier coefficients
for n in range(1, n_terms + 1):
    an[n-1] = (2 / len(t)) * np.sum(y * np.cos(2 * np.pi * n * t / time_end))
    bn[n-1] = (2 / len(t)) * np.sum(y * np.sin(2 * np.pi * n * t / time_end))

# Create the frequency array for the spectrum
frequencies = np.linspace(0, sampling_rate / 2, int(len(t) / 2))

# Calculate the magnitude of the frequency spectrum using Fourier coefficients
Y_magnitude = np.zeros_like(frequencies)
for n in range(1, n_terms + 1):
    # The nth harmonic corresponds to the nth frequency component
    if n < len(frequencies):
        Y_magnitude[n] = np.sqrt(an[n-1]**2 + bn[n-1]**2)

# Plot the frequency spectrum
plt.figure(figsize=(10, 5))
plt.stem(frequencies, Y_magnitude, 'b', markerfmt=" ", basefmt="-b")
plt.title('Frequency Spectrum using Fourier Transform')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic frequency domain data
np.random.seed(0)
frequencies = np.linspace(0, 100, 500)
amplitudes = np.random.normal(1, 0.5, 500) + np.sin(frequencies / 10) * 5

# Define a helper function to plot and save graphs
def plot_feature(frequencies, values, title, filename):
    plt.figure(figsize=(10, 6))
    plt.plot(frequencies, values, label=title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.savefig(f'/mnt/data/{filename}.png')
    plt.close()

# 1. Mean Frequency
mean_frequency = np.mean(amplitudes)
plot_feature(frequencies, amplitudes, f'Mean Frequency: {mean_frequency:.2f}', 'mean_frequency')

# 2. Variance
variance = np.var(amplitudes)
plot_feature(frequencies, amplitudes, f'Variance: {variance:.2f}', 'variance')

# 3. Skewness
skewness = np.mean((amplitudes - np.mean(amplitudes))**3) / np.std(amplitudes)**3
plot_feature(frequencies, amplitudes, f'Skewness: {skewness:.2f}', 'skewness')

# 4. Kurtosis
kurtosis = np.mean((amplitudes - np.mean(amplitudes))**4) / np.std(amplitudes)**4
plot_feature(frequencies, amplitudes, f'Kurtosis: {kurtosis:.2f}', 'kurtosis')

# 5. Frequency Center
frequency_center = np.sum(frequencies * amplitudes) / np.sum(amplitudes)
plot_feature(frequencies, amplitudes, f'Frequency Center: {frequency_center:.2f}', 'frequency_center')

# 6. Standard Deviation
std_dev = np.std(amplitudes)
plot_feature(frequencies, amplitudes, f'Standard Deviation: {std_dev:.2f}', 'standard_deviation')

# 7. Root Mean Square Frequency
rms_frequency = np.sqrt(np.mean(frequencies**2 * amplitudes))
plot_feature(frequencies, amplitudes, f'Root Mean Square Frequency: {rms_frequency:.2f}', 'rms_frequency')

# 8. Spectral Flatness
spectral_flatness = np.exp(np.mean(np.log(amplitudes))) / np.mean(amplitudes)
plot_feature(frequencies, amplitudes, f'Spectral Flatness: {spectral_flatness:.2f}', 'spectral_flatness')

# 9. Energy Distribution
energy_distribution = np.sum(amplitudes**2) / len(amplitudes)
plot_feature(frequencies, amplitudes, f'Energy Distribution: {energy_distribution:.2f}', 'energy_distribution')

# 10. Relative Spread
relative_spread = std_dev / mean_frequency
plot_feature(frequencies, amplitudes, f'Relative Spread: {relative_spread:.2f}', 'relative_spread')

# 11. Weighted Skewness
weighted_skewness = np.sum((frequencies - frequency_center)**3 * amplitudes) / len(frequencies) / std_dev**3
plot_feature(frequencies, amplitudes, f'Weighted Skewness: {weighted_skewness:.2f}', 'weighted_skewness')

# 12. Weighted Kurtosis
weighted_kurtosis = np.sum((frequencies - frequency_center)**4 * amplitudes) / len(frequencies) / std_dev**4
plot_feature(frequencies, amplitudes, f'Weighted Kurtosis: {weighted_kurtosis:.2f}', 'weighted_kurtosis')

# 13. Half-Power Frequency Deviation
half_power_freq_deviation = np.sum((frequencies - frequency_center)**0.5 * amplitudes) / len(frequencies) / np.sqrt(std_dev)
plot_feature(frequencies, amplitudes, f'Half-Power Frequency Deviation: {half_power_freq_deviation:.2f}', 'half_power_frequency_deviation')

# 14. RMS Deviation
rms_deviation = np.sqrt(np.mean((frequencies - frequency_center)**2 * amplitudes))
plot_feature(frequencies, amplitudes, f'RMS Deviation: {rms_deviation:.2f}', 'rms_deviation')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the square wave signal
def square_wave(t, T):
    return np.sign(np.sin(2 * np.pi * t / T))

# Parameters
T = 2 * np.pi  # Period of the square wave
t = np.linspace(0, 2 * T, 1000)  # Time vector
num_terms = 10  # Number of terms in the Fourier series

# Compute the Fourier series
def fourier_series_square_wave(t, T, num_terms):
    x_fs = np.zeros_like(t)
    for n in range(1, 2*num_terms, 2):
        x_fs += (4 / (n * np.pi)) * np.sin(n * 2 * np.pi * t / T)
    return x_fs

# Compute the square wave and its Fourier series approximation
x_square = square_wave(t, T)
x_fs = fourier_series_square_wave(t, T, num_terms)

# Plot the square wave and its Fourier series approximation
plt.figure(figsize=(10, 6))
plt.plot(t, x_square, label='Square Wave', linewidth=2)
plt.plot(t, x_fs, label=f'Fourier Series Approximation (N={num_terms})', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Fourier Series Approximation of a Square Wave')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sine wave signal
def sine_wave(t, frequency):
    return np.sin(2 * np.pi * frequency * t)

# Fourier Series approximation (for sine wave, this is redundant but for demonstration)
def fourier_series_sine_wave(t, frequency, num_terms):
    # Since a sine wave is itself a Fourier component, the Fourier series for sine wave is trivial
    # For demonstration, we'll consider the fundamental frequency component only
    return sine_wave(t, frequency)

# Parameters
frequency = 1  # Frequency of the sine wave in Hz
T = 1 / frequency  # Period of the sine wave
t = np.linspace(0, 2 * T, 1000)  # Time vector (2 periods)
num_terms = 10  # Number of terms in the Fourier series (redundant for pure sine wave)

# Compute the sine wave and its Fourier series approximation
x_sine = sine_wave(t, frequency)
x_fs = fourier_series_sine_wave(t, frequency, num_terms)

# Plot the sine wave and its Fourier series approximation
plt.figure(figsize=(12, 6))

# Original sine wave
plt.subplot(1, 2, 1)
plt.plot(t, x_sine, label='Sine Wave', color='blue')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Sine Wave')
plt.legend()
plt.grid(True)

# Fourier series approximation
plt.subplot(1, 2, 2)
plt.plot(t, x_fs, label=f'Fourier Series Approximation (N={num_terms})', color='red', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Fourier Series Approximation of Sine Wave')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sine wave signal
def sine_wave(t, frequency):
    return np.sin(2 * np.pi * frequency * t)

# Compute the DFT of a signal
def compute_dft(signal):
    return np.fft.fft(signal)

# Parameters
frequency = 5  # Frequency of the sine wave in Hz
sampling_rate = 100  # Sampling rate in Hz
T = 1 / frequency  # Period of the sine wave
t = np.linspace(0, 2 * T, 2 * sampling_rate)  # Time vector (2 periods)

# Generate the sine wave
x_sine = sine_wave(t, frequency)

# Compute the DFT of the sine wave
X_dft = compute_dft(x_sine)
frequencies = np.fft.fftfreq(len(X_dft), d=1/sampling_rate)

# Plot the original sine wave
plt.figure(figsize=(10, 4))
plt.plot(t, x_sine, label='Sine Wave', color='blue')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Original Sine Wave')
plt.legend()
plt.grid(True)
plt.savefig('original_sine_wave_dft.png')
plt.close()

# Plot the magnitude spectrum of the DFT
plt.figure(figsize=(10, 4))
plt.stem(frequencies, np.abs(X_dft), 'r', markerfmt=" ", basefmt="-b")
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.title('DFT Magnitude Spectrum')
plt.grid(True)
plt.savefig('dft_magnitude_spectrum.png')
plt.close()

print("Plots have been saved as 'original_sine_wave_dft.png' and 'dft_magnitude_spectrum.png'")
```
